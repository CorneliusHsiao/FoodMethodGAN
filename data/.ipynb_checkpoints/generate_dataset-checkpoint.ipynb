```python
import os
import json
import numpy
```

```python
all_data_path = "/home/kaijie/Desktop/recipe1M_layers/all_data.json"
with open(all_data_path,'r') as f:
    all_data = json.load(f)
print(len(all_data))
print(all_data[0])
```

```python
ingrs_dict = {}
instrs_dict = {}
part_dict = {}
for d in all_data:
    ingrs = d['ingredients']
    for ingr in ingrs:
        if ingr in ingrs_dict:
            ingrs_dict[ingr]+=1
        else:
            ingrs_dict[ingr]=1
        
    instrs = d['instructions']
    for instr in instrs:
        if instr in instrs_dict:
            instrs_dict[instr]+=1
        else:
            instrs_dict[instr]=1
    
    part = d['partition']
    if part in part_dict:
        part_dict[part]+=1
    else:
        part_dict[part]=1
    
print(len(ingrs_dict))
print(len(instrs_dict))
print((part_dict))
```

```python
count = 0
for k,v in ingrs_dict.items():
    if k+"s" in ingrs_dict:
        ingrs_dict[k+"s"]+=1
        count+=1

print(count)
print(len(ingrs_dict))

new_ingrs_dict = {}
count = 0
for k,v in ingrs_dict.items():
    if not k+"s" in ingrs_dict:
        new_ingrs_dict[k] = v
    else:
        count+=1
print(count)
print(len(new_ingrs_dict))


ingrs_dict = new_ingrs_dict
print(len(ingrs_dict))
count=0
new_ingrs_dict = {}
for k,v in ingrs_dict.items():
    if  v > 100:
        count+=1
        new_ingrs_dict[k] = v
ingrs_dict = new_ingrs_dict
print(count)
print(len(ingrs_dict))

ingrs_dict = new_ingrs_dict.copy()
print(ingrs_dict)
```

```python
new_all_data = []
for i,d in enumerate(all_data):
    ingrs = d['ingredients']
    new_ingrs = []
    for j in ingrs:
        if j=="":
            continue
        elif j in ingrs_dict:
            new_ingrs.append(j)
        elif j+"s" in ingrs_dict:
            new_ingrs.append(j+"s")
            
    new_d = d.copy()
    new_d['ingredients'] = new_ingrs
    new_all_data.append(new_d)
```
