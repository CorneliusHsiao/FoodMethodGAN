```python
import os
import json
import numpy
```

```python
all_data_path = "/home/kaijie/Desktop/recipe1M_layers/all_data.json"
with open(all_data_path,'r') as f:
    all_data = json.load(f)
print(len(all_data))
print(all_data[0])
```

```python
ingrs_dict = {}
instrs_dict = {}
part_dict = {}
for d in all_data:
    ingrs = d['ingredients']
    for ingr in ingrs:
        if ingr in ingrs_dict:
            ingrs_dict[ingr]+=1
        else:
            ingrs_dict[ingr]=1
        
    instrs = d['instructions']
    for instr in instrs:
        if instr in instrs_dict:
            instrs_dict[instr]+=1
        else:
            instrs_dict[instr]=1
    
    part = d['partition']
    if part in part_dict:
        part_dict[part]+=1
    else:
        part_dict[part]=1
    
print(len(ingrs_dict))
print(len(instrs_dict))
print((part_dict))
```

```python
count = 0
for k,v in ingrs_dict.items():
    if k+"s" in ingrs_dict:
        ingrs_dict[k+"s"]+=1
        count+=1

print(count)
print(len(ingrs_dict))

new_ingrs_dict = {}
count = 0
for k,v in ingrs_dict.items():
    if not k+"s" in ingrs_dict:
        new_ingrs_dict[k] = v
    else:
        count+=1
print(count)
print(len(new_ingrs_dict))


ingrs_dict = new_ingrs_dict
print(len(ingrs_dict))
count=0
new_ingrs_dict = {}
for k,v in ingrs_dict.items():
    if  v > 100:
        count+=1
        new_ingrs_dict[k] = v
ingrs_dict = new_ingrs_dict
print(count)
print(len(ingrs_dict))

ingrs_dict = new_ingrs_dict.copy()
print(ingrs_dict)
```

```python
new_all_data_path = "/home/kaijie/Desktop/recipe1MDataset/data_all.json"
with open(new_all_data_path,'r') as f:
    new_data = json.load(f)

print(len(new_data))
```

```python
train_num = 0
val_num = 0
test_num = 0
for d in new_data:
    part = d['partition']
    if part=="train":
        train_num+=1
    elif part=="val":
        val_num+=1
    else:
        test_num+=1

total = train_num+val_num+test_num
print(total)
print(train_num/total,val_num/total,test_num/total)
```

```python
train_data=[]
val_data=[]
test_data=[]

for d in new_data:
    part = d['partition']
    if part=="train":
        train_data.append(d.copy())
    elif part=="val":
        val_data.append(d.copy())
    else:
        test_data.append(d.copy())

print(len(train_data),len(val_data),len(test_data))
```

```python
datas = [train_data,val_data,test_data]
paths = ["train.json",'val.json','test.json']
root_path = "/home/kaijie/Desktop/recipe1MDataset/"

for i,d in enumerate(datas):
    path = os.path.join(root_path,paths[i])
    with open(path,'w',encoding='utf8') as f:
        str_ = json.dumps(d,
                          indent=4, sort_keys=True,
                          separators=(',', ': '), ensure_ascii=False)
        f.write(to_unicode(str_))
```
